```python
#import lib
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
class normal_method:
    def __init__(self, a1, b1, c1):
        self.a1 = a1
        self.b1 = b1
        self.c1 = c1
        
    def dist_curve(self):
        plt.plot(self.c1, 1/(self.b1 * np.sqrt(2 * np.pi)) *
            np.exp( - (self.c1 - self.a1)**2 / (2 * self.b1**2) ), linewidth=2, color='y')
        plt.show()
```

```python
#Vary the mean and SD to generate different plots
mean_ = 5 
sd_ = 2
```

```python
c = np.random.normal(mean_, sd_, 3000)
```

```python
w1, x1, z1 = plt.hist(c, 100, normed=True) #hist
hist = normal_method(mean_, sd_, x1)
plot = hist.dist_curve()
```
