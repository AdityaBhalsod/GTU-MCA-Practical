```python
#import lib
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
xs = np.arange(-10, 10, 0.001)
x2s = [x**2 for x in xs]
```

```python
for k in [1, 4, 8, 16, 24]:
    plt.plot(x2s, chi2.pdf(x2s, k), label='k = {0}'.format(k))

plt.xlim(0, 40)
plt.ylim(0, 0.5)
plt.legend()
plt.grid()
plt.show()
```
