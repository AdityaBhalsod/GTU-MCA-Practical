```python
#import lib
import numpy as np
import matplotlib.pyplot as plt
import warnings
from scipy.misc import factorial
warnings.filterwarnings('ignore')
```

```python
t = np.arange(0, 20, 0.1)
d = np.exp(-5)*np.power(5, t)/factorial(t)
```

```python
plt.plot(t, d, 'bs')
plt.grid()
plt.show()
```
